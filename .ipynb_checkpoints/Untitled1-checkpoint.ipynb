```python
df = pd.read_csv("BEM_CANDIDATO_2010_2006.csv")
```

```python
df["VALOR_BEM"] = df["VALOR_BEM"].str.replace('.', '')
```

```python
df["VALOR_BEM"] = df["VALOR_BEM"].str.replace(',', '.')
```

```python
df["VALOR_BEM"] = df["VALOR_BEM"].astype(float)
df["ANO_ELEICAO"] = df["ANO_ELEICAO"].astype(float)
df["NOME_CANDIDATO"] = df["NOME_CANDIDATO"].astype(str)
# df["SIGLA_PARTIDO"].astype(str)
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE"').groupby(["ANO_ELEICAO", "NOME_CANDIDATO"]).sample()
```

```python
print("Comparação dos bens dos candidatos em 2006")
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE" and ANO_ELEICAO == 2006').groupby("NOME_CANDIDATO").sum("VALOR_BEM").head()
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE" and ANO_ELEICAO == 2006').groupby("NOME_CANDIDATO").sum("VALOR_BEM").plot(x='NUMERO_CANDIDATO', y="VALOR_BEM", kind="bar", title="Comparação dos bens dos candidatos em 2006")
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE" and ANO_ELEICAO == 2010').groupby("NOME_CANDIDATO").sum("VALOR_BEM").plot(x='NUMERO_CANDIDATO', y="VALOR_BEM", kind="bar", title="Comparação dos bens dos candidatos em 2010")
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE"').groupby(["ANO_ELEICAO", "NOME_CANDIDATO"]).sum("VALOR_BEM")
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE" and ANO_ELEICAO == 2006')["SIGLA_PARTIDO"].value_counts().plot(x='SIGLA_PARTIDO', y="", kind="bar", title="Quantidade de candidatos no partido em 2006")
```

```python
df.query('DESCRICAO_CARGO  == "PRESIDENTE" and ANO_ELEICAO == 2010')["SIGLA_PARTIDO"].value_counts()
```
